```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Downloads/student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
# change weekly study hours column
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","10-20")
df.head()
```

```python
# GENDER DISTRIBUTION
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x = "Gender")
plt.title("Gender distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# from the above char we have analysed that the number of female data is more than the number of male data 
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot=True)
plt.title("parent education")
plt.show()
```

```python
# from the above chart we have concluded that the education of the parents have a good impact in their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot=True)
plt.title("Students score relation between parent marital status ")
plt.show()
```

```python
# from the above chart we have concluded that there is no/ negligible impact on the students score due to their parents marital status
```
